```python
import os

from modules.feature_engineering_function import *
from modules.eda_functions import *
from modules.data_collection_functions import *
```

```python
path = '../data/historical/'
```

```python
all_data_as_dict = import_all_files_as_dict(path)
```

```python
for company in all_data_as_dict.keys():
    all_data_as_dict[company] = all_data_as_dict[company].sort_index()
```

```python
all_data_as_dict
```

```python
companies = all_data_as_dict.keys()
```

```python
for company in companies:
    all_data_as_dict[company] = all_data_as_dict[company].drop(columns=['Adj Close', 'High', 'Low', 'Open', 'Volume'])
```

```python
all_data_as_dict
```

```python
for company in companies:
    all_data_as_dict[company] = create_lag(all_data_as_dict[company], 'Close', 1)
```

```python
all_data_as_dict
```

```python
for company in companies:
    close = all_data_as_dict[company]['Close']
    lag_1 = all_data_as_dict[company]['lag_1']
    labels = [1 if close[i] < lag_1[i] else 0 for i in range(len(close))]
    all_data_as_dict[company]['label'] = labels
```

```python
try:
    os.mkdir('../data/lags')
except:
    print('Already Exists')
finally:
    path = '../data/lags/'
    export_to_csv(all_data_as_dict, path, companies)
```
