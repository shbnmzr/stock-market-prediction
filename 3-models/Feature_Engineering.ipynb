```python
import os

import numpy as np
import pandas as pd

from modules.feature_engineering_function import *
from modules.eda_functions import *
from modules.data_collection_functions import *
```

```python
path = '../data/historical/'
```

```python
all_data_as_dict = import_all_files_as_dict(path)
```

```python
for company in all_data_as_dict.keys():
    all_data_as_dict[company] = all_data_as_dict[company].sort_index()
```

```python
all_data_as_dict
```

```python
companies = all_data_as_dict.keys()
```

```python
for company in companies:
    all_data_as_dict[company] = all_data_as_dict[company].drop(columns=['Adj Close', 'High', 'Low', 'Open', 'Volume'])
```

```python
all_data_as_dict
```

```python
normalized = {}
for company in all_data_as_dict.keys():
    X = np.array(all_data_as_dict[company]['Close'])
    normalized[company] = pd.DataFrame(z_score_normalization(X), index=all_data_as_dict[company].index, columns=['Normalized Close'])
```

```python
normalized
```

```python
for company in companies:
    for i in range(5):
        normalized[company] = create_lag(normalized[company], 'Normalized Close', i+1)
```

```python
normalized
```

```python
for company in companies:
    close = normalized[company]['Normalized Close']
    lag_1 = normalized[company]['lag_1']
    labels = [0 if close[i] < lag_1[i] else 1 for i in range(len(close))]
    normalized[company]['label'] = labels
```
