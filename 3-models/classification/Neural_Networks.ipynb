```python
from modules.models.neural_networks import *
from modules.models.modelling_functions import *
from modules.eda_functions import *
from tensorflow.python.keras.activations import linear, relu
from sklearn.metrics import confusion_matrix, classification_report
```

```python
path = '../../data/classification-data/'
```

```python
all_data_as_dict = import_all_files_as_dict(path)
```

```python
all_data_as_dict
```

```python
layers = layers_construction(10, [90, 80, 70, 60, 50, 40, 30, 20, 10, 2], ['relu', 'relu', 'relu', 'relu', 'relu', 'relu', 'relu', 'relu', 'relu', 'linear'])
```

```python
models = {}

for company in all_data_as_dict.keys():
    X, y = all_data_as_dict[company].drop(['label'], axis=1), all_data_as_dict[company]['label']
    X_train, y_train, X_dev, y_dev, X_test, y_test = split_train_dev_test(X, y, train_size=0.7, test_size=0.15)
    model = model_construction(layers)
    compile_and_fit(model, X_train, y_train, 50)
    y_hat = make_inference(model, X_dev)

    models[company] = {
        'model': model,
        'y_hat': y_hat,
        'X_dev': X_dev,
        'y_dev': y_dev,
        'X_test': X_test,
        'y_test': y_test
    }
```

```python
scores = {}
for company in all_data_as_dict.keys():
    score = evaluate_model(models[company]['y_hat'], models[company]['y_dev'])
    scores[company] = score
```

```python
scores
```

```python
predictions = {}
for company in all_data_as_dict.keys():
    y_hat = make_inference(models[company]['model'], models[company]['X_test'])
    predictions[company] = y_hat
```

```python
for company in all_data_as_dict.keys():
    print(f"Confusion Matrix {company}:\n{confusion_matrix(models[company]['y_test'], predictions[company])}")
```

```python
for company in all_data_as_dict.keys():
    print(f"Report: {company}:\n{classification_report(models[company]['y_test'], predictions[company], zero_division=0)}")
```

```python
for company in predictions.keys():
    predictions[company] = {
        'True Values': models[company]['y_test'],
        'Predictions': predictions[company]
    }
```

```python
pred_path = '../../data/predictions/neural-networks-cls'
export_predictions(predictions, pred_path)
```
